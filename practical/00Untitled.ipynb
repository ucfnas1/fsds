```python
#Using the example from "Who's on the List?" in Lecture 2.3 
#(and feel free to refer to the video), 
#briefly explain why you might choose index over in when searching for content in a list.
geographers = ["Massey","Harvey","Rose"]
print("Massey" in geographers) # True
print("Batty" in geographers)  # False
```

in only checks whether an item appears in a list, returning True or False, but it doesn’t show where the item is. index gives the exact position, which is useful for accessing related information. However, index will raise an error if the item is missing, so it must be used more carefully.
