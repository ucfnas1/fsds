# Model Question 2
Using the example from "Who's on the List?" in Lecture 2.3 (and feel free to refer to the video), briefly explain why you might choose index over in when searching for content in a list.
when searching for content in a list. 

Answer2:
When searching a list, sometimes you need more than just checking whether an item exists.
The in operator only returns True or False and does not tell you where the item is.
The index method returns the exact position (an integer) of the item, although it raises an error if the item is not in the list.

Or:
When searching a list, "in" only tells you whether an item exists, so the program continues running even if the value is missing.
"index", on the other hand, is useful when you need to ensure the item is present: if it cannot be found, it raises a ValueError, which intentionally stops execution.
So you would choose index when the absence of the item should be treated as a failure rather than something the program silently ignores.

```python
# 找元素（整数字符）
grographers = ['Massy', 'Harvey', 'Rose']
print(grographers[0])
print(grographers[-1])
print(grographers[1.5])  #整数字符
```

```python
# slice切片
print(grographers[0:])
print(grographers[0:1])
print(grographers[0:2])
print(grographers[0:3])
print(grographers[:])
print(grographers[:3])

print(grographers[1:2])
print(grographers[1])
```

```python
# .index() 找元素的位置
grographers = ['Massy', 'Harvey', 'Rose']
grographers.index('Massy')
```

```python
# .index() 有位置，找元素是谁
grographers = ['Massy', 'Harvey', 'Rose']
print(grographers[grographers.index('Massy')])  #先找元素的位置，再找元素是谁
```

```python
# .sort() 首字母排序
grographers = ['Massy', 'Harvey', 'Rose']
grographers.sort()  #按照首字母（从小到大）排序
print(grographers)

grographers.sort(reverse=True)  #按倒序（Z → A）
print(grographers)

grographers.sort(reverse=False) #按顺序（Z → A）
print(grographers)
```

```python
# change list 改变元素
grographers = ['Massy', 'Harvey', 'Rose']
grographers[2] = 'Jefferson'
print(grographers)
```

```python
# .insert() 插入元素  指明(位置，元素)
grographers = ['Massy', 'Harvey', 'Rose']
grographers.insert(0,'von Humboldt')
print(grographers)

# .pop() 删除元素  直接指明(位置)
grographers.pop(3) 
print(grographers)
```

```python
# 合并（内容）两个List 用加法+
male_grographers = ['A', 'B', 'C']
female_grographers = ['D', 'E', 'F']
all_grographers = male_grographers + female_grographers

print(all_grographers)
print(all_grographers[0])

# 连接两个List 或者用 .extend()
male_grographers = ['A', 'B', 'C']
female_grographers = ['D', 'E', 'F']
male_grographers.extend(female_grographers)  #向左合并内容
print(male_grographers)
```

```python
# 连接两个List 用.append()
male_grographers = ['A', 'B', 'C']
female_grographers = ['D', 'E', 'F']
all_grographers = []  #清空旧内容，创建一个新的列表//否则会在原list的基础上append
all_grographers.append(male_grographers)
all_grographers.append(female_grographers)

print(all_grographers)
print(all_grographers[0])
print(all_grographers[-1])

# append 的对象是 list → 结果就会是嵌套 list
```
